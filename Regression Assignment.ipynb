# Polynomial Regression

#### Consider synthetic data for a noisy sine wave

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
X = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
y = np.sin(X) + np.random.normal(scale=0.1, size=X.shape)

# Plot the data
plt.scatter(X, y, color='blue', label='Noisy sine wave')
plt.title('Synthetic Data: Noisy Sine Wave')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

### Fit a polynomial regression model of degree 3 to the synthetic data and visualize the results.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Transform the data to include polynomial features
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
```

```python
# Fit the model
poly_reg_model = LinearRegression()
poly_reg_model.fit(X_poly, y)
```

```python
# Predict using the model
y_pred = poly_reg_model.predict(X_poly)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', label='Noisy sine wave')
plt.plot(X, y_pred, color='red', label='Polynomial regression (degree=3)')
plt.title('Polynomial Regression (degree=3)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Ridge and Lasso Regression

### Consider simple dataset 

```python
# Generate a simple linear dataset
X_simple = np.random.rand(1000, 1) * 10
y_simple = 4 * X_simple ** 2 - X_simple 

# Plot the data
plt.scatter(X_simple, y_simple, color='blue', label='Data')
plt.title('Simple Linear Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

### Fit Ridge and Lasso regression models to the simple dataset and visualize the results.

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_simple, y_simple, test_size=0.2, random_state=42)
```

```python
# Fit Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
```

```python
# Fit Lasso Regression
lasso = Lasso(alpha=0.1, max_iter=10000)
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
```

```python
# Plot the results
plt.scatter(X_test, y_test, color='blue', label='Data')
plt.plot(X_test, y_pred_ridge, color='red', label='Ridge regression')
plt.plot(X_test, y_pred_lasso, color='green', label='Lasso regression')
plt.title('Ridge and Lasso Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
